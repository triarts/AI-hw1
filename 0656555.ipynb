```python
#class for contain 1 vertex only - id vertex id edge, id edge, cost
# masih di pertimbangkan enaknya taruh edge as obj atau id doang tp ntar ga btuh graph klo taruh obj
class Vertex(object):
    def __init__(self, idVertex, edge1Id, edge2Id, cost):
        self.idVertex = idVertex;
        self.edge1 = edge1Id;
        self.edge2 = edge2Id;
        self.cost = cost;
        
    def getId(self):
        return self.idVertex;
        
    def getCost(self):
        return self.cost;
    
    # return edge 1 id
    def getEdge1Id(self):
        return self.edge1;
    
    # return edge 2 id
    def getEdge2Id(self):
        return self.edge2;
    
    def setCost(self, cost):
        self.cost = cost;
```

```python
# class for contain list of vertex
class Vertexes(object):
    def __init__(self, listVertex):
        self.listVertex = listVertex;
        
    def length(self):
        return len(self.listVertex);
    
    # return vertex with search by id vertex
    def getVertexById(self, vertexId):
        for obj in self.listVertex:
            if obj.getId() == vertexId:
                return obj;
        return 0;
    
    # return all vertex id as list
    def getAllVertexId(self):
        listId = [];
        for obj in self.listVertex:
            listId.append(obj.getId());
        return listId;
    
    # get all vertex as list
    def getVertexes(self):
        return self.listVertex;
```

```python
# calss for graph and some function to retrieve data from vertexex and Edges
# ntar di liat vertex enakana nyatuke edge(object) atau edge id tok
class Graph(object):
    def __init__(self, mEdges, mVertexes):
        self.mEdges = mEdges;
        self.mVertexes = mVertexes;
        
    ## ==vertex function
    # search all vertex id that contain node id 'x'
    def getAllVertexIdContainEdgeId(self, edgeId):
        tempListVertexId = [];
        for obj in self.mVertexes.getVertexes():
            if obj.getEdge1Id() == edgeId:
                tempListVertexId.append(obj.getId());
            elif obj.getEdge2Id() == edgeId:  
                tempListVertexId.append(obj.getId());
        return tempListVertexId;
    
    # get specific vertex contain EdgeId 'a' && 'b'
    def getVertexContainEdgeId(self, EdgeId1,EdgeId2):
        #tempListVertex = [];
        for obj in self.mVertexes.getVertexes():
            if obj.getEdge1Id() == EdgeId1 and obj.getEdge2Id() == EdgeId2:
                #tempListVertex.append(obj);
                return obj;
            elif obj.getEdge1Id() == EdgeId2 and obj.getEdge2Id() == EdgeId1:
                return obj
        return None;
    
    # Extended from vertexes class
    def getVertexById(self, vertexId):
        return self.mVertexes.getVertexById(vertexId);
    
    # Extended from vertexes class
    def getAllVertexId(self):
        return self.mVertexes.getAllVertexId();
    
    
    ## === edges function
    # search all neighbour edge id connected to edge id 'x'
    def getAllNeighbourEdgeIdByEdgeId(self,nodeId):
        tempListEdgeId = [];
        for obj in self.mVertexes.getVertexes():
            if obj.getEdge1Id() == nodeId:
                tempListEdgeId.append(obj.getEdge2Id());
            elif obj.getEdge2Id() == nodeId:
                tempListEdgeId.append(obj.getEdge1Id());
            #print(tempListNodeId);
        return tempListEdgeId;
    
    # Extented from Edges class
    def getEdgeById(self, edgeId):
        return self.mEdges.getEdgeById(edgeId);
    
    def getAllEdgeId(self):
        return self.mEdges.getAllEdgeId()
    
    ## == common function
    def getEdges(self):
        return self.mEdges;
    def getVertexes(self):
        return self.mVertexes;
    
    def EdgesLen(self):
        return self.mEdges.length();
    
    def VertexesLen(self):
        return self.mVertexes.length();
    
```

```python
# container for astar variable a.k.a route from start to goal
# contain H,C,F (F = H+C)
# C in here is accumulative cost
class AStarNode(object):
    def __init__(self, newEdge,graph):
        self.H = 0;
        self.C = 0;
        self.F = self.H+self.C;
        self.EdgeList = [];
        self.addEdge(newEdge,graph);
        
    #==getter
    def getHval(self):
        return self.H;
    def getCval(self):
        return self.C;
    def getFval(self):
        return self.H + self.C;
    
    #==setter
    def setHval(self, H):
        self.H = H;
    def setFval(self, F):
        self.F = F;
    def setCval(self, C):
        self.C = C;
        
    #calculate total cost
    def calcTotalCval(self, graph):
        tempCost = 0;
        if len(self.EdgeList) == 1:
            tempCost = 0;
        elif len(self.EdgeList) > 1:
            for i in range(len(self.EdgeList) - 1):
                tempCost=tempCost + graph.getVertexContainEdgeId(self.EdgeList[i].getId(),self.EdgeList[i+1].getId()).getCost();
        self.setCval(tempCost);
        
    #calcualte F
    def calcFval(self):
        self.setFval(self.C + self.H);
        
    #calculate H
    def calcHval(self):
        self.setHval( self.getEndEdge().getHval() );
    
    # == Edge management ==
    # add new Edge id to path
    def addEdge(self,Edge,graph):
        self.EdgeList.append(Edge);
        self.calcTotalCval(graph);
        self.calcHval();
        self.calcFval();
        
    # get start Edge for sorting and remove purpose
    def getStartEdge(self):
        if len(self.EdgeList) != 0:
            return self.EdgeList[0];
        else:
            return None
        
    # get end Edge for sorting and remove purpose | next step purpose
    def getEndEdge(self):
        if len(self.EdgeList) != 0:
            return self.EdgeList[len(self.EdgeList)-1];
        else:
            return None
    
    # return Edge(obj) as list
    def getEdgePath(self):
        return self.EdgeList;
    
    #return only path of id edge
    def getEdgePathId(self):
        EdgeIdList = [];
        for obj in self.EdgeList:
            EdgeIdList.append(obj.getId())
        return EdgeIdList;
    
    # calculate path count eq: 'abc' = 3, abcd = 4
    def pathCount(self):
        return len(self.getEdgePathId());
    
    #calculate the weight of id path, a = 97, b= 98, dst
    def pathPriority(self):
        pathW = 0;
        tempPathId = self.getEdgePathId();
        for i in range (0,self.pathCount()):
            pathW = pathW + ord(tempPathId[i])
        return pathW;
    
    def printPathId(self):
        print(self.getEdgePathId())
    #cek apakah ada cycle di pathnya apa ndak
    # true ada cycle 
    # false tidak ada cycle
    def isCyclePath(self):
        unique = set(self.getEdgePathId())  
        for each in unique:  
            count = self.getEdgePathId().count(each)  
            if count > 1:  
                #print 'There are duplicates in this list'  
                return True  
        #print 'There are no duplicates in this list'  
        return False  
    
    
```

```python
# special class of priority queue with sort based on F value, for A star algo
# ## ========== nodelist contain list of 'AStarNode' =============
class PriorityQueue(object):
    def __init__(self):
        self.nodeList = [];
        #self.nodeList = node
        
    # == basic queue function ==
    # insert with rule
    def superInsert(self, astarNode):
        if not astarNode.isCyclePath():
            self.nodeList.append(astarNode)
        else:
            print("node have cycle -- dropped")
     #normal basic insert       
    def insert(self, astarNode):
        self.nodeList.append(astarNode)
    
    #remove /clean queue with specific rule
    # remove if same path have F value is bigger, keep only the smallest one
    def cleanUp(self):
        rmList = []
        # awal semua ga di hapus
        for i in range (len(self.nodeList)):
            rmList.append(0);
        
        # loop untuk nandain idx mana yang perlu di hapus
        for i in range (len(self.nodeList)): # kok kyak e hrus e ada -1 nya ya
            # cek kalau belum kena remove flag
            if rmList[i] == 0 :
                for y in range (i+1,len(self.nodeList)):
                    # cek kalau start dan endnya sama
                    if self.isStartEndNodeSame(self.nodeList[i],self.nodeList[y]):
                        #sementara compare pake value total path
                        res = self.compareFval(self.nodeList[i],self.nodeList[y]);
                        # cek hasil comparasi Fvalue
                        if(res == 1 or res == 3):
                            rmList[i] = 1;
                            break
                        elif (res == 2 or res == 4):
                            rmList[y] = 1;
#                         else:
#                             rmList[y] = 1;
        
        #get index dr rmList        
        rmIdx = [];
        for i in range (len(rmList)):
            if rmList[i] == 1:
                rmIdx.append(i);
                
        # remove from queue
        for index in sorted(rmIdx, reverse=True):
            self.remove(index)
                    
    # remove from queue
    def remove(self, idx):
        self.nodeList.pop(idx)
        
    # get and remove from queue
    def get(self, idx):
        return self.nodeList.pop(idx)
    # peek
    def peek(self,idx=None):
        if idx is not None:
            # return specific index
            return self.nodeList[idx];
        else:
            #return last one
            return self.nodeList[self.count()-1];
    
    
    def count(self):
        return len(self.nodeList);
        
    # 1 = 1 > 2
    # 2 = 1 < 2
    # 3 = F1 == F2 && len(1) > len(2)
    # 4 = F1 == F2 && len(1) < len(2)
    def compareFval(self,ASN1,ASN2):
        if ASN1.getFval() > ASN2.getFval():
            return 1;
        elif ASN1.getFval() < ASN2.getFval():
            return 2;
        elif len(ASN1.getEdgePath()) > len(ASN2.getEdgePath()):
            return 3;
        elif len(ASN1.getEdgePath()) < len(ASN2.getEdgePath()):
            return 4;
#         elif  ASN1.getFval() == ASN2.getFval() and len(ASN1.getEdgePath()) == len(ASN2.getEdgePath()):
#             return 5;
        
    # cek untuk apakah start Edge dan End edge sama atau beda | untuk remove dr list
    # return true or false
    def isStartEndNodeSame(self, astarNode1, astarNode2):
        an1ListId = astarNode1.getEdgePathId();
        an2ListId = astarNode2.getEdgePathId();
        for i in range(0,len(an1ListId)):
            for y in range(len(an2ListId)-1,len(an2ListId)):
                if an1ListId[i] == an2ListId[y]:
                    return True;
        return False;
#         if astarNode1.getStartEdge().getId() == astarNode2.getStartEdge().getId():
#             #cek id path 1-1
#             an1ListId = astarNode1.getEdgePathId();
#             an2ListId = astarNode2.getEdgePathId();
#             for i in range(1,len(an1ListId)):
#                 for y in range(len(an2ListId)-1,len(an2ListId)):
#                     if an1ListId[i] == an2ListId[y]:
#                         return True;
#             return False;
#         else:
#             return False;
    # main sort function based on Fvalue
    def sort(self):
        #sort by F value
        self.nodeList.sort(key=lambda x: x.getFval());
        
    # sort for handling same F value path and path length
    def sortByPathCount(self):
        self.nodeList.sort(key=lambda x: x.pathCount());
    
    # sortint based on Path pirotiy from astar node class
    def sortByPathPriority(self):
        #sort by Path priority
        self.nodeList.sort(key=lambda x: x.pathPriority());
    
    def printQ(self, idx=None):
        tempID = []
        if idx is None:
            for i in range( len (self.nodeList) ):
                tempID.append(self.nodeList[i].getEdgePathId()) # print astar node path
                tempID.append(self.nodeList[i].getFval()) # print F value of astar node
            print (tempID)
        else:
            tempID.append(self.nodeList[idx].getEdgePathId()) # print astar node path
            tempID.append(self.nodeList[idx].getFval()) # print F value of astar node
            print("printQ idx")
            print(tempID)
    
    # complement function
    def convert(self, list): 
        # Converting integer list to string list 
        # and joining the list using join() 
        res = str("".join(map(str, list))) 
        return res
    
    #return queue path  as list
    def getQueueAsList(self,idx=None):
        tempList = [];
        if idx is None:
            for i in range( len (self.nodeList) ):
                tempList.append(self.convert( self.nodeList[i].getEdgePathId() ) ) # print astar node path
        else:
            tempList.append(self.convert(self.nodeList[idx].getEdgePathId()) ) # astar node path
        return tempList; 
    
```

```python
class Astar(object):
    def __init__(self, graph):
        self.graph = graph;
        self.PQexplore = PriorityQueue();
        self.startEdgeId = '';
        self.endEdgeId = '';
        self.FinalPath = [];
    
    def isGoal(self, EdgeId):
        if self.endEdgeId == EdgeId:
            return True
        else:
            return False;
    
    def getPQ(self):
        return self.PQexplore;
    
    def computeFinalPath(self):
        #self.FinalPath.pop(0);
        self.FinalPath.insert(0,self.startEdgeId);
            
    def getFinalPathHistory(self):
        return self.FinalPath;
    
    def readFile(self,fileName):
        retVal=[]
        f = open(fileName, "r")
        retVal = f.read().splitlines()
        return retVal;

    def writeFile(self,fileName,content):
        f = open(fileName, "w")
        for x in content:
            f.write(str(x)+"\n")

        
    def startSearchPath (self,EdgeId_start, EdgeId_goal):
        self.startEdgeId = EdgeId_start;
        self.endEdgeId = EdgeId_goal;
        
        
        startNode = AStarNode( self.graph.getEdgeById(self.startEdgeId),self.graph )
        self.PQexplore.insert(startNode); # main Priority Queue
        
        #selama hasilnya blm 1 rute dan endnodenya belum goal node
        loop_count = 0;
        while self.PQexplore.count() != 1 or self.PQexplore.peek(0).getEndEdge().getId() != EdgeId_goal:
            
            # debug first state of queue
            print("queue first state");
            self.PQexplore.printQ() 
            
            neighbourIdList = []; # list for contain neigbour edge id from parent node
            tempPQ = PriorityQueue(); # temporary Priority Queue for explore purpose
            
            # get data idx [0] karena asusmsi sudah di sort dan yang paling kecil di index[0]
            # cek misal end node same as goal then explore next node
            exploreIdx = 0
            flagFind = False; # cek apakah idx 0 sdh sampe goal apa blm
            for i in range(0,self.PQexplore.count()):
                if self.PQexplore.peek(exploreIdx).getEndEdge().getId() == self.endEdgeId:
                    if i == 0:
                        flagFind = True;
                    else:
                        exploreIdx = exploreIdx+1;
                else:
                    break
            if flagFind is True:
                self.PQexplore.printQ(0)
                self.FinalPath.append(self.PQexplore.getQueueAsList(0)) # put into final list
                break
                
            # cek misal ada 2 atau lebih path yg sdh sampe goal tp kembar F valuenya
            print("pq count: "+str(self.PQexplore.count()) +" expIdx : "+str(exploreIdx))
            if(self.PQexplore.count() > 1 and self.PQexplore.count() == exploreIdx):
                print("log : at special case")
                # cek dlu apakah ada node length path nya beda
                flag = False;
                for i in range(0,self.PQexplore.count()-1):
                    for y in range(i+1,self.PQexplore.count()):
                        if self.PQexplore.peek(i).pathCount() == self.PQexplore.peek(y).pathCount():
                            flag = True;
                            
                if flag is True:
                #kalau ada yang beda len nya brarti sort dr len, hasil akhir idx[0]
                    self.PQexplore.sortByPathCount();
                    self.PQexplore.printQ(0)
                    self.FinalPath.append(self.PQexplore.getQueueAsList(0)) # put into final list
                else:
                    #sort dr abcnya, ambil idx[0]
                    self.PQexplore.sortByPathPriority();
                    self.PQexplore.printQ(0)
                    self.FinalPath.append(self.PQexplore.getQueueAsList(0)) # put into final list
                break
            else:
                #list all neighbour realted to that node
                neighbourIdList = self.graph.getAllNeighbourEdgeIdByEdgeId( self.PQexplore.peek(exploreIdx).getEndEdge().getId() )
                print("parent : "+self.PQexplore.peek(exploreIdx).getEndEdge().getId())
                print (neighbourIdList)

                parentNode = self.PQexplore.get(exploreIdx); # node that need to explore
                tempNode = None; # for astarnode tamporary variable

                # start explore neighbour node
                for y in range( len(neighbourIdList) ):
                    # di insert sudah sma ceknya
                    tempNode = copy.deepcopy(parentNode);
                    #tempNode = parentNode;
                    tempNode.addEdge( self.graph.getEdgeById(neighbourIdList[y]), self.graph )
                    print("combined path : ")
                    tempNode.printPathId()
                    tempPQ.superInsert(tempNode)
                    print("new add into queue : ")
                    tempPQ.printQ()

                # put back into original queue
                for i in range (0,tempPQ.count()):
                    tempPQNode = copy.deepcopy(tempPQ.peek(i));
                    self.PQexplore.insert(tempPQNode)

                # sort queue
                self.PQexplore.sort();
                print("sorted queue : ")
                self.PQexplore.printQ()
                self.FinalPath.append(self.PQexplore.getQueueAsList()) # put into final list

                # remove dr queue
                self.PQexplore.cleanUp();
                print("after clean up queue : ")
                self.PQexplore.printQ()
                self.FinalPath.append(self.PQexplore.getQueueAsList()) # put into final list
            
            # for debug and control purpose
            loop_count=loop_count+1;
            print ("==================== lc : "+str(loop_count))
            if loop_count == 1000:
                print("weird")
                break
            
        print("loop count : "+str(loop_count) );
            
```
