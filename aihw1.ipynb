```python
# class for contain 1 edge only
class Edge(object):
    def __init__(self, nodeId, Hvalue):
        self.nodeId = nodeId;
        self.Hvalue = Hvalue;

    def getId(self):
        return self.nodeId;

    def getHval(self):
        return self.Hvalue;
    
    def setHval(self,hval):
        self.Hvalue = hval;
    
    
```

```python
# class for contain list of edge
class Edges(object):
    def __init__(self, listNode):
        self.listNode = [];
        self.listNode = listNode;

    def length(self):
        return len(self.listNode);

    # return node with search by id node
    def getEdgeById(self, nodeId):
        for obj in self.listNode:
            if obj.getId() == nodeId:
                return obj;
        return 0;
    
    def getAllEdgeId(self):
        listId = [];
        for obj in self.listNode:
            listId.append(obj.getId());
        return listId;
    
    def getListNode(self):
        return self.listNode;
```

```python
#class for contain 1 vertex only - id vertex id edge, id edge, cost
# masih di pertimbangkan enaknya taruh edge as obj atau id doang tp ntar ga btuh graph klo taruh obj
class Vertex(object):
    def __init__(self, idVertex, edge1Id, edge2Id, cost):
        self.idVertex = idVertex;
        self.edge1 = edge1Id;
        self.edge2 = edge2Id;
        self.cost = cost;
        
    def getId(self):
        return self.idVertex;
        
    def getCost(self):
        return self.cost;
    
    # return edge 1 id
    def getEdge1Id(self):
        return self.edge1;
    
    # return edge 2 id
    def getEdge2Id(self):
        return self.edge2;
    
    def setCost(self, cost):
        self.cost = cost;
```

```python
# class for contain list of vertex
class Vertexes(object):
    def __init__(self, listVertex):
        self.listVertex = listVertex;
        
    def length(self):
        return len(self.listVertex);
    
    # return vertex with search by id vertex
    def getVertexById(self, vertexId):
        for obj in self.listVertex:
            if obj.getId() == vertexId:
                return obj;
        return 0;
    
    # return all vertex id as list
    def getAllVertexId(self):
        listId = [];
        for obj in self.listVertex:
            listId.append(obj.getId());
        return listId;
    
    # get all vertex as list
    def getVertexes(self):
        return self.listVertex;
```

```python
# calss for graph and some function to retrieve data from vertexex and Edges
# ntar di liat vertex enakana nyatuke edge(object) atau edge id tok
class Graph(object):
    def __init__(self, mEdges, mVertexes):
        self.mEdges = mEdges;
        self.mVertexes = mVertexes;
        
    ## ==vertex function
    # search all vertex id that contain node id 'x'
    def getAllVertexIdContainEdgeId(self, edgeId):
        tempListVertexId = [];
        for obj in self.mVertexes.getVertexes():
            if obj.getEdge1Id() == edgeId:
                tempListVertexId.append(obj.getId());
            elif obj.getEdge2Id() == edgeId:  
                tempListVertexId.append(obj.getId());
        return tempListVertexId;
    
    # get specific vertex contain EdgeId 'a' && 'b'
    def getVertexContainEdgeId(self, EdgeId1,EdgeId2):
        tempListVertex = [];
        for obj in self.mVertexes.getVertexes():
            if obj.getEdge1Id() == EdgeId1 and obj.getEdge2Id() == EdgeId2:
                tempListVertex.append(obj);
            elif obj.getEdge1Id() == EdgeId2 and obj.getEdge2Id() == EdgeId1:
                tempListVertex.append(obj);
        return tempListVertex;
    
    # Extended from vertexes class
    def getVertexById(self, vertexId):
        return self.mVertexes.getVertexById(vertexId);
    
    # Extended from vertexes class
    def getAllVertexId(self):
        return self.mVertexes.getAllVertexId();
    
    
    ## === edges function
    # search all neighbour edge id connected to edge id 'x'
    def getAllNeighbourNodeIdByEdgeId(self,nodeId):
        tempListEdgeId = [];
        for obj in self.mVertexes.getVertexes():
            if obj.getEdge1Id() == nodeId:
                tempListEdgeId.append(obj.getEdge2Id());
            elif obj.getEdge2Id() == nodeId:
                tempListEdgeId.append(obj.getEdge1Id());
            #print(tempListNodeId);
        return tempListEdgeId;
    
    # Extented from Edges class
    def getEdgeById(self, edgeId):
        return self.mEdges.getEdgeById(edgeId);
    
    def getAllEdgeId(self):
        return self.mEdges.getAllEdgeId()
    
    ## == common function
    def getEdges(self):
        return self.mEdges;
    def getVertexes(self):
        return self.mVertexes;
    
    def EdgesLen(self):
        return self.mEdges.length();
    
    def VertexesLen(self):
        return self.mVertexes.length();
    
```

```python
# container for astar variable
# contain H,C,F (F = H+C)
class AStarNode(object):
    def __init__(self, H,C):
        self.H = H;
        self.C = C;
        self.F = self.H+self.C;
        self.idList = [];
        
    #getter
    def getHval(self):
        return self.H;
    def getCval(self):
        return self.C;
    def getFval(self):
        return self.H + self.C;
    
    #setter
    def setHval(self, H):
        self.H = H;
    def setCval(self, C):
        self.C = C;
    def setFval(self):
        self.F = self.C + self.H;
    
    # == Edge id management ==
    # add new Edge id to path
    def addEdgeId(self,EdgeId):
        self.idList.append(EdgeId);
    # get start Edge id for sorting and remove purpose
    def getStartEdgeId(self):
        if len(self.idList) != 0:
            return self.idList[0];
        else:
            return None
    # get end Edge id for sroting and remove purpose
    def getEndEdgeId(self):
        if len(self.idList) != 0:
            return self.idList[len(self.idList)-1];
        else:
            return None
    
    
```
